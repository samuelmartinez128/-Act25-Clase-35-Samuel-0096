```python
# --- PROGRAMA POO: CLASE USUARIO ---

class Usuario:
    # El método __init__ inicializa los 3 atributos solicitados
    def __init__(self):
        self.nombre = ""
        self.contrasena = ""
        self.rol = ""

    # Función 1: Realiza la entrada de datos por teclado
    def leerdatos(self):
        print(f"--- Ingresando datos del usuario ---")
        self.nombre = input("Nombre de usuario: ")
        self.contrasena = input("Contraseña: ")
        self.rol = input("Rol (Admin/Cliente/Invitado): ")
        print("Datos guardados con éxito.\n")

    # Función 2: Muestra la información almacenada (Contraseña visible)
    def mostrardatos(self):
        print(f"--- Perfil de Usuario ---")
        print(f"Usuario: {self.nombre}")
        print(f"Contraseña: {self.contrasena}")
        print(f"Rol: {self.rol}")
        print("-" * 25)

# --- INSTANCIACIÓN (Creación de los 2 objetos) ---

# Creamos el primer objeto y leemos sus datos
usuario1 = Usuario()
usuario1.leerdatos()

# Creamos el segundo objeto y leemos sus datos
usuario2 = Usuario()
usuario2.leerdatos()

# --- LLAMADAS A LAS FUNCIONES PARA MOSTRAR RESULTADOS ---

print("REVISIÓN DE USUARIOS REGISTRADOS:")
usuario1.mostrardatos()
usuario2.mostrardatos()

print("Programa terminado")
print("Gestión de Usuarios POO - Adaptado del modelo de Samuel Martinez")
```
